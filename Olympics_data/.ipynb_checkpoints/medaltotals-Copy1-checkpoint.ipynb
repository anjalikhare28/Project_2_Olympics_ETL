```python
#%config Completer.use_jedi = False


import pandas as pd
from sqlalchemy import create_engine
#from config import *
```

```python
ath_2016_file = "Resources/athletes_2016.csv"
ath_2016_df = pd.read_csv(ath_2016_file)
ath_2016_df.head()
```

```python
clean_ath_2016_df = ath_2016_df[['name', 'nationality','sport']]
clean_ath_2016_df
```

```python
clean_ath_2016_df.describe()
```

```python
top_athlete = clean_ath_2016_df[clean_ath_2016_df['name'] == 'Jianan Wang']
top_athlete
```

```python
clean_ath_2016_df = ath_2016_df[['name', 'nationality', 'sex', 'sport', 'gold', 'silver', 'bronze']]
clean_ath_2016_df
```

```python
countries_2016_file = "Resources/countries_2016.csv"
country_2016_df = pd.read_csv(countries_2016_file)
country_2016_df.head()
```

```python
clean_country_df = country_2016_df.rename(columns = {'code':'nationality'})
clean_country_df
```

```python
merged_2016_df = pd.merge(clean_ath_2016_df,clean_country_df,  on="nationality", how="left")
merged_2016_df
```

```python
merged_2016_df = merged_2016_df[['name', 'country', 'sport', 'gold', 'silver', 'bronze' ]]
merged_2016_df
```

```python
#merged_2016_df.dropna(inplace=True)
#merged_2016_df
```

```python
medal_2016 = merged_2016_df.groupby(['country'])[['gold', 'silver', 'bronze']].sum()
medal_2016['total2016'] = medal_2016.sum(axis=1)
medal_2016.dropna(how = 'all')
medal_2016
```

```python
ath_2021_file = "Resources/athletes_2021.csv"
ath_2021_df = pd.read_csv(ath_2021_file)
clean_ath_2021_df = ath_2021_df.rename(columns={'Name':'name','NOC':'country','Discipline':'sport'})
clean_ath_2021_df.head()
```

```python
clean_ath_2021_df.describe()
```

```python
top_athlete_2021 = clean_ath_2021_df[clean_ath_2021_df['name'] == 'KURBANOV Ruslan']
top_athlete_2021
```

```python
medals_2021_file = "Resources/medals_2021.csv"
medals_2021_df = pd.read_csv(medals_2021_file)
medals_2021_df.head()
```

```python
medals_2021_df.rename(columns={'Team/NOC':'Country'}, inplace=True)
medals_2021_df = medals_2021_df[['Country','Gold','Silver','Bronze','Total']]
medals_2021_df.rename(columns={'Country':'country','Gold':'gold','Silver':'silver','Bronze':'bronze','Total':'total'},inplace=True)
medals_2021_df.sort_values('total', inplace=True, ascending=False)
medals_2021_df.head()
```

```python
merged_medals = pd.merge(medal_2016, medals_2021_df, on="country", how="outer")
merged_medals
```

```python
cleaned_merged_medals = merged_medals[['country', 'total2016', 'total']]
cleaned_merged_medals
```

```python
cleaned_merged_medals = cleaned_merged_medals.rename(columns = {'total2016':'medals2016','total':'medals2021'})
cleaned_merged_medals.head()
```

```python
cleaned_merged_medals.dropna(inplace=True)
cleaned_merged_medals
```

```python
merged_2021_df = pd.merge(clean_ath_2021_df,medals_2021_df,  on="country", how="left")
merged_2021_df
```
