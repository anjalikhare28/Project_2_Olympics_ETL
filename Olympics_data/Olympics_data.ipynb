```python
%config Completer.use_jedi = False


import pandas as pd
from sqlalchemy import create_engine
from config import *
```

```python
ath_2016_file = "Resources/athletes_2016.csv"
ath_2016_df = pd.read_csv(ath_2016_file)
ath_2016_df.head()
```

```python
countries_2016_file = "Resources/countries_2016.csv"
country_2016_df = pd.read_csv(countries_2016_file)
country_2016_df.head()
```

```python
ath_2021_file = "Resources/athletes_2021.csv"
ath_2021_df = pd.read_csv(ath_2021_file)
ath_2021_df.head()
```

```python
medals_2021_file = "Resources/medals_2021.csv"
medals_2021_df = pd.read_csv(medals_2021_file)
medals_2021_df.head()
```

```python
clean_ath_2016_df = ath_2016_df[['name', 'nationality', 'sex', 'sport', 'gold', 'silver', 'bronze']]
clean_ath_2016_df
```

```python
clean_country_df = country_2016_df.rename(columns = {'code':'nationality'})
clean_country_df
```

```python
merged_2016_df = pd.merge(clean_ath_2016_df, clean_country_df, on="nationality", how="left")
merged_2016_df
```

```python
merged_2016_df = merged_2016_df[['name', 'country', 'sport', 'gold', 'silver', 'bronze' ]]
merged_2016_df
```

```python
merged_2016_df.dropna(inplace=True)
merged_2016_df
```

```python
medal_2016 = merged_2016_df.groupby(['country'])[['gold', 'silver', 'bronze']].sum()
medal_2016
```

```python
medal_2016.loc['India'] 
```

```python
medal_2016 ['medals'] = medal_2016['gold'] +  medal_2016['silver'] +  medal_2016['bronze']
medal_2016.sort_values('medals', inplace=True, ascending=False)
medal_2016
```
